```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter dasar
Ac = 1        # Amplitudo sinyal pembawa (Carrier amplitude)
Am = 3     # Amplitudo sinyal informasi (Message amplitude)
fc = 200     # Frekuensi sinyal pembawa (Carrier frequency) dalam Hz
fm = 50       # Frekuensi sinyal informasi (Message frequency) dalam Hz
mu = Am / Ac  # Indeks modulasi

# Frekuensi sampling dan waktu sampling
Fs = 1000  # Frekuensi sampling (Sampling frequency) dalam Hz
t = np.arange(0, 1, 1/Fs)  # Waktu dari 0 hingga 0.05 detik

# Sinyal informasi (Message signal)
m_t = Am * np.cos(2 * np.pi * fm * t)

# Sinyal pembawa (Carrier signal)
c_t = Ac * np.cos(2 * np.pi * fc * t)

# Sinyal AM dengan indeks modulasi
s_t = Ac * (1 + mu * np.cos(2 * np.pi * fm * t)) * np.cos(2 * np.pi * fc * t)

# Plot sinyal informasi, sinyal pembawa, dan sinyal AM
plt.figure(figsize=(10, 8))

# Plot sinyal informasi (message signal)
plt.subplot(3, 1, 1)
plt.plot(t, m_t)
plt.title('Sinyal Informasi (m(t))')
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')
plt.grid(True)

# Plot sinyal pembawa (carrier signal)
plt.subplot(3, 1, 2)
plt.plot(t, c_t)
plt.title('Sinyal Pembawa (c(t))')
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')
plt.grid(True)

# Plot sinyal yang telah dimodulasi (AM signal)
plt.subplot(3, 1, 3)
plt.plot(t, s_t)
plt.title(f'Sinyal AM dengan Indeks Modulasi  = {mu:.2f}')
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')
plt.grid(True)

# Tampilkan plot
plt.tight_layout()
plt.show()
```
